```python
def find_nonzero(x):
    #return the indices of non-zero elements of an array.

    #Input:      x:  array
    #Output:     indices: array of integer indices
    indices = np.where(x !=0)
    return np.array(indices).T

y4 = np.array([[1,2],[0,0],[4,0]])
y4 = np.array([1,2,0,0,4,0])
x4 = find_nonzero(y4)
x4
```

Q12. Creating a new array

```python
def cart2polar(xy):
    #radius and angle (unit: radii)
    xyData = np.array(xy).T
    radius = np.sqrt(xyData[0]**2 + xyData[1]**2)
    polar = np.arctan(xyData[1]/xyData[0])
    return np.array([radius, polar]).T

xyTest = [[1, 1], [1, 2]]
print(cart2polar(xyTest))  
```

Q14. Compute and time for loops and array operations.

```python
def calculate(a, b):
    """Do some calculations
    
    Take in two arrays, subtract 1/4 of the [:2,:] elements from the elements themselves.
    Then multiply with array b and return the result.
    
    Inputs:
        a: input array, shape [MxN]
        b: input array, shape [M]
    
    Returns:
        resultant array
    """

    aa = a[:2, :].copy() #Change: add .copy() to avoid array alias
    aa /= 4.
    a[:2, :] -= aa
    return a*b
    

a = np.ones((3, 2))
b = np.ones((3,1))*3 #Change: the shape of b has change from (3,) to (3,1), 
                     #for the array broadcast operation Generally, all of the dimensions need to be either the same size, 
                     #or one of the dimension sizes for a particular dimension should be of size 1. 
res = calculate(a, b)
assert (res == np.array([[ 2.25,  2.25],[ 2.25,  2.25],[ 3., 3.]])).all()
```

Q15. Fix the code

```python
import numpy as np

data = np.loadtxt('./burl1h2010.txt')
#print(data[:,5:7])

def vector_components(data):
    rad = np.pi/180
    components = []
    for wdir, wspd in data:
        u = wspd * np.cos(wdir*rad)
        v = wspd * np.sin(wdir*rad)
        components.append([u,v])
    return components

vector_components(data[:,5:7])
```

Q16. Calculate components of a vector

```python
import glob
class dataclass:
    def __init__(self,filename):
        self.name = filename
        temp = []
        salt = []
        with open(self.name) as f:
            for line in f:
                #print(line.split())
                spline = line.split()
                if '*END*' in spline:
                    break
            for line in f:
                spline = line.split()
                temp.append(float(spline[2]))
                salt.append(float(spline[5]))
        self.temperature = temp
        self.salinity = salt
    def salt_mean(self):
        return np.mean(self.salinity)
    def temp_max(self):
        return np.max(self.temperature)

def show():

    file_list = glob.glob('./MCH/*.cnv2')
    wind = []
    for file in file_list:
        wind.append(dataclass(file))
    #show wind salt_mean() temp_max()
    for w in wind:
        print(w.salt_mean())
        print(w.temp_max())
show()
        
```

Q17. (Grad) Write and use a class

```python
def running_mean(vec,N):
    length = len(vec)
    #truncate near the end
    if N >= 0:
        return [np.mean(vec[i:i+N]) if i < length - N else np.mean(vec[i:]) for i in range(length)]
    else:
        return [np.mean(vec[:i+1]) if i < -N else np.mean(vec[i+N:i]) for i in range(length)]
   
Z = list(range(1,100))
N = -3
print(running_mean(Z,-3))
print(running_mean(Z,3))
```
